```python
import urllib.request
import time

PAPERS = [
    {"id": "1706.03762", "filename": "attention_is_all_you_need.pdf", "title": "Attention Is All You Need"},
    {"id": "1606.06565", "filename": "concrete_problems_ai_safety.pdf", "title": "Concrete Problems in AI Safety"},
    {"id": "2212.08073", "filename": "constitutional_ai.pdf", "title": "Constitutional AI"},
    {"id": "2201.03544", "filename": "reward_hacking_survey.pdf", "title": "Reward Hacking Survey"},
    {"id": "1706.03741", "filename": "rlhf_christiano.pdf", "title": "Deep RL from Human Preferences"},
    {"id": "1805.00899", "filename": "scalable_oversight_debate.pdf", "title": "AI Safety via Debate"},
]

for paper in PAPERS:
    dest = f"data/papers/{paper['filename']}"
    
    if os.path.exists(dest):
        print(f"[skip] {paper['title']} — already exists")
        continue
    
    url = f"https://arxiv.org/pdf/{paper['id']}.pdf"
    print(f"[downloading] {paper['title']} ...")
    
    urllib.request.urlretrieve(url, dest)
    size_kb = os.path.getsize(dest) // 1024
    print(f"[ok] {size_kb} KB saved")
    
    time.sleep(1)  # be polite to arxiv

print("\ndone!")
```
